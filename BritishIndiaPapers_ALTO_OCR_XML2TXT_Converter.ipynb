# XML to TXT Converter

We can convert all ALTO XML files to plain text files using this script. One thing we will need to do is to keep all XML files in the /files directory.

```python
# Some code has been adopted from https://github.com/cneud/alto-ocr-text/blob/master/alto_ocr_text.py

import codecs
import os
import sys
import xml.etree.ElementTree as ET
```

```python
directory = "./files"
files = os.listdir(directory)
print(files)
dsstore = '.DS_Store'
if dsstore in files: files.remove(dsstore)
content = ''
for filepath in files:
    parser=ET.XMLParser(encoding='iso-8859-5')
    tree = ET.parse(directory+'/'+filepath)
#     txtFileName = filepath[:-3]+'txt'
    txtFileName = 'result.txt'
    xmlns = tree.getroot().tag.split('}')[0].strip('{')
    for lines in tree.iterfind('.//{%s}TextLine' % xmlns):
        content = content+'\n'
        for line in lines.findall('{%s}String' % xmlns):
            text = line.attrib.get('CONTENT') + ' '
            content = content+text
ft = open(directory+'/'+txtFileName, "w+")
ft.write(content)
ft.close()
```
