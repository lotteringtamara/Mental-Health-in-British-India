## Basic Cleaning

```python
def countWords(string, wordsToCount):
    string = string.lower() # Text normalization: make string lowercase
    string = re.sub(r'[^\w\s]','', string) # Text normalization: remove punctuation
    splitString = string.split() # Split string into array of words
    counts = Counter(splitString) # Get counts for each word like Counter({'dogs': 3, 'cute': 1})
    count = 0 # Start the counter
    for word in wordsToCount: # Loop through list of words and add the count
        count = count + counts[word]
    return count

def cleanOnWordCount(country, tokens, root):
    print("Processing", country, "...")
    # Directory declarations
    myCorpusRoot = str(root) + str(country)
    myCorpusRootClean = str(root) + str(country) + 'Clean3'
    # Create a directory if it doesn't exist
    if not os.path.exists(myCorpusRootClean):
        os.makedirs(myCorpusRootClean) 
    filteredFiles = 0
    for filename in os.listdir(myCorpusRoot):
        if filename.endswith(".txt"):
            curFilePath = os.path.join(myCorpusRoot, filename)
            curFileOpen = open(curFilePath, "r")
            curFileText = curFileOpen.read()
            curFileCount = countWords(curFileText, tokens)
            if(curFileCount > 2): # Word count for tokens (sum) should be greater than 2
                filteredFiles = filteredFiles + 1
                copyfile(curFilePath, str(myCorpusRootClean)+'/'+str(filename))
    print('Copied', filteredFiles, 'to the new directory for the country', country)
```

```python
def getCountryDataFrame(country, tokens, root):
    print("Working...")
    myCorpusRootClean = str(root) + str(country) + 'Clean3'
    filteredFiles = 0
    dfList = []
    for filename in os.listdir(myCorpusRootClean):
        if filename.endswith(".txt"):
            dfListLib = {}
            curFilePath = os.path.join(myCorpusRootClean, filename)
            curFileOpen = open(curFilePath, "r")
            curFileText = curFileOpen.read()
            curFileCount = countWords(curFileText, tokens)
            dfListLib['country'] = country
            dfListLib['file'] = filename
            dfListLib['file_path'] = curFilePath
            dfListLib['token_freq'] = curFileCount
            dfListLib['date_yearmonth'] = filename[0:6]
            dfListLib['date_year'] = filename[0:4]
            dfListLib['date_month'] = filename[4:6]
            dfListLib['date_day'] = filename[6:8]
            dfListLib['text'] = curFileText
            dfList.append(dfListLib)
    return dfList
```

```python
def cleanTextInDf(mystring):
    mystring = mystring.lower() # Text normalization: make string lowercase
    mystring = re.sub(r'[^\w\s]','', mystring) # Text normalization: remove punctuation
    return mystring
```

```python
# Get nouns and adjectives from the text

from nltk import word_tokenize, pos_tag

def nounsAdj(text):
    isNounAdj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nounsAdj = [word for (word, pos) in pos_tag(tokenized) if isNounAdj(pos)] 
    return ' '.join(nounsAdj)
```

```python
data_nouns_adj = pd.DataFrame(dfCleanGermany.text.apply(nounsAdj))
data_nouns_adj
```

```python
vectorizerNounAdj = CountVectorizer(stop_words='english')
dataVectorizerNA = vectorizerNounAdj.fit_transform(data_nouns_adj.text)
dataDtmGermanyNA = pd.DataFrame(dataVectorizerNA.toarray(), columns = vectorizerNounAdj.get_feature_names())
dataDtmGermanyNA.index = dataDtmGermanyNA.index # lets get the indices from previous dataframe
dataDtmGermanyNA # This is the document term matrix
```

```python
corpusNA = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(dataDtmGermanyNA.transpose()))

id2wordNA = dict((v, k) for k, v in vectorizerNounAdj.vocabulary_.items())
```

```python
ldaNA = models.LdaModel(corpus=corpusNA, num_topics=4, id2word=id2wordNA, passes=10)
ldaNA.print_topics()
```

```python
ldaNA = models.LdaModel(corpus=corpusNA, num_topics=4, id2word=id2wordNA, passes=80)
ldaNA.print_topics()
```
