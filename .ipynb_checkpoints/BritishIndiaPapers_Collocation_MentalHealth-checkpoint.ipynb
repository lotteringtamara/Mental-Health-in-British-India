This notebook uses the clean and filtered data to analyse the n-grams in the corpus.

```python
import nltk # Using NLTK for ngrams analysis
from nltk.util import ngrams
from gensim.parsing.preprocessing import remove_stopwords
from sklearn.feature_extraction.text import CountVectorizer
import os
import pandas as pd
import collections
import re
import seaborn as sns # For graphs

sw = stopwords.words("english")
```

```python
df = pd.read_csv('df.csv')
del df['Unnamed: 0']
df
```

```python
def formatText(text):
    text = ''.join([i for i in text if not i.isdigit()])
    text = " ".join(text.split())
    text = remove_stopwords(text)
    text = ' '.join( [w for w in text.split() if len(w)>1] )
    return text

cleanText = lambda text: formatText(text) # Lambda function applies to all cells in a column
cleandf = pd.DataFrame(df.text.apply(cleanText)) # .apply() the function to all cells
df = cleandf
```

```python
df
```

```python
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:100]
```

### Bigrams

```python
top_bi_grams = get_top_ngram(df['text'],n=2)
x,y = map(list,zip(*top_bi_grams))
sns.set(rc={'figure.figsize':(14,30)})
sns.barplot(x=y,y=x)
```

### Trigrams

```python
top_tri_grams = get_top_ngram(df['text'],n=3)
x,y = map(list,zip(*top_tri_grams))
sns.set(rc={'figure.figsize':(14,30)})
sns.barplot(x=y,y=x)
```

### 4-grams

```python
top_quad_grams = get_top_ngram(df['text'],n=4)
x,y = map(list,zip(*top_quad_grams))
sns.set(rc={'figure.figsize':(14,30)})
sns.barplot(x=y,y=x)
```
